```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# % passing math: score of 70% or greater
math_approved_df = (school_data_complete.loc[school_data_complete['math_score'] >= 70].math_score.count()) / student_count * 100
math_approved_df
```

```python
# % passing reading: scores of 70% or greater
reading_approved_df = (school_data_complete.loc[school_data_complete['reading_score'] >= 70].reading_score.count()) / student_count * 100
reading_approved_df
```

```python
school_data_complete.to_excel("pycityschool.xls", sheet_name="Sheet1")
```

```python
# % Overall Passing: students passing both math and reading 70% or higher
#student_count passing both math & reading
overall_pass_df = (reading_approved_df + math_approved_df) / 2
```

```python
overall_pass_df
```

```python
#Create a Dataframe to hold the above results.Title: School Summary
District_Summary = District_Summary = pd.DataFrame({"Total Schools":[school_count], "Total Studens":[student_count], "Total Budget":[Total_budget], "Average Math Score":[ave_math], "Average Reading Score":[ave_read], "% Passing Math":[math_approved_df], "% Passing Reading":[reading_approved_df], "% Overall Passing Rate":[overall_pass_df]})
```
